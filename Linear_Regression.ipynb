```python
from sklearn.datasets import make_regression
```

```python
x, y = make_regression(n_samples=400, n_features=1, noise=43, bias=1)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
x = x.ravel()
```

```python
class LinearRegression:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.m = 0
        self.c = 0
        self.n = len(x)
        
    def fit(self, epochs, lr):
        # gradient descent
        for i in range(epochs):
            y_pred = self.m*self.x + self.c
            
            # derivative
            dm = (2/self.n)*sum((self.x*((y_pred)-self.y)))
            dc = (1/self.n)*sum(((y_pred)-self.y))
            
            # update parameters
            self.m = self.m - lr*dm
            self.c = self.c - lr*dc
            
    def predict(self, inp):
        y_pred = self.m*inp + self.c
        return y_pred
```

```python
reg = LinearRegression(x, y)
```

```python
reg.fit(epochs=500, lr=.001)
```

```python
y_pred = reg.predict(x)
```

```python
#Plotting the results
plt.figure(figsize = (10,6))
plt.scatter(x, y)
plt.plot(x, y_pred , color = 'k' , lw = 3)
plt.xlabel('x' , size = 20)
plt.ylabel('y', size = 20)
plt.show()
```
